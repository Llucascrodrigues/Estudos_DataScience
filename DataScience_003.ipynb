#Análise de dados de filmes
##Arrecadação

```python
import pandas as pd
import seaborn as sns

#vamos analisar quanto o filme arrecadou (revenue = coluna de arrecação)
```

```python
tmdb=pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/tmdb_5000_movies.csv")
tmdb.head()
```

```python
sns.displot(tmdb["revenue"])
'''vemos que o gráfico explode para numeros menores, vamos investigar
e ver o que está acontecendo, criando um filtro para observarmos valores menores
que 500'''
```

```python
# prompt: gráfico de distribuição do orçamento dos filmes (budget)

sns.displot(tmdb["budget"])
'''A IA gerou o grafico do orçamento, conseguimos notar a diferença deste grafico
para com o anterior, fortalecendo que devemos investigar os dados para valores pequenos'''
```

```python
tmdb.query("revenue<500")
#faturamentos menores que 500, não foram registrados por isso está zerado, então vamos filtrar esse
```

```python
com_faturamento=tmdb.query("revenue>0")
sns.displot(com_faturamento["revenue"])
'''Para os filmes com dados coletados de receita esse é o gráfico que mais faz sentido
pois leva em consideração todos os dados lançados de lucros.
podemos ver que realmente filmes arrecadam menos que 500 mil mas agora com um comportamento
que faz mais sentido do que aquela explosão do gráfico anterior'''
```

```python
#vamos investigar a lingua dos filmes, pesquisando os valores únicos
tmdb["original_language"].unique()
```
